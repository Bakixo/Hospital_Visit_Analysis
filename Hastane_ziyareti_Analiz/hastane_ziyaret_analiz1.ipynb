## TRAÄ°N TEST SPLÄ°T

```python
# df.to_csv("community_health_MLR", index=False)
```

```python
df2 = pd.read_csv("community_health_MLR.csv")
```

```python
df2
```

```python
X = df2.drop("visits",axis = 1)
```

```python
y = df2["visits"]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=99)
```

```python
X_train.head()
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train,y_train)
```

```python
test_preds = regr.predict(X_test)
```

```python
test_preds
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print("Theta :     ", round(regr.coef_[0],2))
print("Intercept : ", round(regr.intercept_,2))
print("R^2 :       ", round(r2_score(y_test,test_preds),2))
print("MAE :       ", round(mean_absolute_error(y_test,test_preds),2))
print("RMSE :      ", round(np.sqrt(mean_squared_error(y_test,test_preds)),2)) 

# WOW BAYA Ä°YÄ°
```
